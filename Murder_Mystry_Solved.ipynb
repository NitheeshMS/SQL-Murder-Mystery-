**1.  Retrieve Crime Scene Report:**

**Task:** Run a query to retrieve the crime scene report for the murder that occurred on Jan.15, 2018, in SQL City. Gather all available details from the report.

```python
import sqlite3
# Establish a connection to the database
conn = sqlite3.connect('/content/drive/MyDrive/SQL_Murder n Webscrapping/sql-murder-mystery.db')

# Create a cursor object to execute SQL queries
cursor = conn.cursor()

# Run a query to retrieve the crime scene report for the specified murder
query = '''
SELECT *
FROM crime_scene_report
WHERE date = '20180115' AND type = 'murder' AND city = 'SQL City'
'''

cursor.execute(query)

# Fetch the result
result = cursor.fetchone()

# Print the details
if result:
    print("Crime Scene Report:")
    print(f"Date: {result[0]}")
    print(f"Type: {result[1]}")
    print(f"Description: {result[2]}")
    print(f"City: {result[3]}")
else:
    print("Crime scene report not found for the specified criteria.")

```

**2. Witness Personal Details:**

**Task:**Check the personal details of witnesses involved in the case. Retrieve their names, addresses, and any other relevant information.


```python
import sqlite3

# Establish a connection to the database
connection = sqlite3.connect('/content/drive/MyDrive/SQL_Murder n Webscrapping/sql-murder-mystery.db')

# Create a cursor object to execute SQL queries
cursor = connection.cursor()

# SQL query to retrieve witness details
query = """
    SELECT p.name AS witness_name, p.address_number, p.address_street_name, p.ssn
    FROM person p
    JOIN interview i ON p.id = i.person_id
    WHERE i.transcript LIKE '%Murder%'; -- Assuming 'Murder' is mentioned in the transcript
"""

# Execute the query and fetch the result
cursor.execute(query)
witness_details = cursor.fetchall()

# Print or process the retrieved witness details
for row in witness_details:
    print("Witness Name:", row[0])
    print("Address:", row[1], row[2])
    print("Social Security Number:", row[3])
    print("\n")

# Don't forget to close the cursor and the connection when done
cursor.close()
connection.close()

```
